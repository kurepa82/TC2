### Transferencia

Realizamos el análisis por el método sistemático de **nodos.**

 En el primer nodo nos queda:
 
$$
\begin{equation}
\label{eqn:nudo1}
V_1(s) \cdot (G_a+G+G) - V_A(s) \cdot G_a = 0
\end{equation}
$$
 En el segundo nodo nos queda:
 $$
\begin{equation}
\label{eqn:nudo2}
-V_1(s) \cdot G - V_B(s) \cdot G_k = 0
\end{equation}
$$
+ Trabajando con las ecuaciones que se obtuvieron de los nodos y sabiendo que 

$$ \begin{equation}
\label{eqn:v}
VAB(s) = V_A(s)-V_B(s)
\end{equation}
$$
$$
\begin{equation}
\label{eqn:dife}
VAB(s) = V_1(s) \cdot 2 \cdot k
\end{equation}
$$

Y obtenemos la salida diferencial pedida y observamos que es un amplificador en configuración _**inversora**. Para ver el desarrollo en papel,ingrese aquí.
