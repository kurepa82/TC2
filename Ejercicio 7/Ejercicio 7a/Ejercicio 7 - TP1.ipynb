Se plantean las ecuaciones en las entradas del **OPAMP**

+ Analizando la _entrada inversora_
\begin{equation}
\label{eqn:inv}
\frac{V_1(s)-V_a(s)}{R_1} + \frac{V_2(s)-V_a(s)}{R_2} = 0
\end{equation}

+ Analizando la _entrada no inversora_
\begin{equation}
\label{eqn:noinv}
(V_1(s)-V_a(s)) \cdot sC - \frac{V_a(s)}{R_3} = 0
\end{equation}

+ Finalmente _combinando_ las ecuaciones obtenemos

\begin{equation}
\label{eqn:trans}
T(s) = \frac{V_2(s)}{V_1(s)} = \frac{s-\frac{R_2}{C \cdot R_1 \cdot R_3}}{s+\frac{1}{C \cdot R_3}} = \frac{s-1000}{s+1000}
\end{equation}

+ Si se desea _normalizar_ la transferencia 

Consideramos que la frecuencias donde se ubican los polos valen 1 r/s

$\omega$$_c$ = $\omega$$_p$ = $\frac{1}{C \cdot R_3}$ = 1 $\frac{r}{s}$

\begin{equation}
\label{eqn:nor}
T(s) = \frac{V_2(s)}{V_1(s)} = \frac{s-1000}{s+1000} = \frac{s-1}{s+1}
\end{equation}

$\frac{V2}{V1}(j\omega)=\frac{j\omega-\frac{R2}{R1}.\omega_o}{j\omega+\omega_o}$

$|\frac{V2}{V1}|²= \frac{\omega²+(\frac{R2}{R1}.\omega_o)²}{\omega²+\omega_o²}$

$\phi= arctg(\frac{\omega}{-\frac{R2}{R1}.\omega_o})-arctg(\frac{\omega}{\omega_o})$
