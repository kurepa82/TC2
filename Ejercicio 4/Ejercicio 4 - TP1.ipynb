Se propone el siguiente circuito para reemplazar la resistencia que está _**conectada a masa**_

\begin{equation}
\label{eqn:imp}
Z_1(s) = \frac{V_1(s)}{I_o(s)}
\end{equation}

\begin{equation}
\label{eqn:tra}
I_o(s) = gm_1 \cdot V_2(s) = \frac{V_1(s) - V_2(s)}{R_1} 
\end{equation}

\begin{equation}
\label{eqn:final}
Z_1(s) = R_1 + \frac{1}{gm_1}
\end{equation}

Emulando nuestro _**resistor serie a masa**_.



Se propone el siguiente circuito para reemplazar a la resistencia que está _**flotando**_.

\begin{equation}
\label{eqn:curr}
I_i(s) = I_o(s) = gm_1 \cdot (V_1(s) - V_o(s) = gm_2 \cdot V_o(s)
\end{equation}

De la ecuación anterior tenemos toda la información necesaria para plantear como se comporta el _**divisor resistivo**_

\begin{equation}
\frac{V_o(s)}{V_1(s)} = \frac{gm_1}{gm_1 + gm_2} 
\end{equation}

\begin{equation}
Z_1(s) = \frac{gm_1 + gm_2}{gm_1 \cdot gm_2} = R_1 + R_2
\end{equation}
